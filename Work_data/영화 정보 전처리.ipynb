# 중복 영화명 제거

```python
movie_df = pd.read_csv('c:/2nd_project/Data/movie_data/2023_0107_merge_mid.csv')
movie_df
```

```python
len(movie_df['영화명'].unique())
```

```python
movie_df[movie_df['영화명'] == '드림']
```

```python
movie_df = movie_df.groupby(['영화명']).agg({
    '개봉일': 'first',
    '등급' : 'first',
    '장르' : 'first',
    '대표국적' : 'first',
    '국적' : 'first',
    '제작사' : 'first',
    '배급사' : 'first',
    '감독' : 'first',
    '배우' : 'first',
}).reset_index()

movie_df
```

```python
movie_df[movie_df['개봉일'] == '2023-08-03']
```

# 23년 1월 1일 ~ 23년 7월 31일 사이 개봉한 영화

```python
from datetime import datetime

# '개봉일' 컬럼을 datetime 형식으로 변환
movie_df['개봉일'] = pd.to_datetime(movie_df['개봉일'])
movie_df['영화명'] = movie_df['영화명'].astype(str)

# '개봉일'이 2023년 1월 1일부터 2023년 7월 31일 사이인 데이터만 선택
movie_df = movie_df[(movie_df['개봉일'] >= datetime(2023, 1, 1)) & (movie_df['개봉일'] <= datetime(2023, 7, 31))]
movie_df
```

# 최종 데이터

```python
movie_df.to_csv('c:/2nd_project/Data/movie_data/2023_01~07_movie.csv', index=None)
```
