```python
def change_ner_data(txts):
    for txt in tqdm(txts):
        first_line = '; ' + txt + '\n'
        main_lines = []
        pos = komoran.pos(txt)

        for i in range(len(pos)):
            # print(pos[i][0])
            for ner_list in ner_lists:
                word = ner_list.split('\t')[0]
                mor = ner_list.split('\t')[1]
                tag = ner_list.split('\t')[2]
                # print(word, mor, tag)
                if ((pos[i][0] == word) & (len(pos[i][0]) == len(word))) & (pos[i][1] == mor):
                    # print('\nner ==============================')
                    # print(f'{ner_list}')
                    # print(f'word : {word}\nmor : {mor}\ntag : {tag}')
                    # print('\ntxt ==============================')
                    # print(f'pos : {pos[i][0]}\ntag : {tag}')
                    bio_tag = tag
                    # main_word = word
                    # main_bio_tag = tag
        #             second_line = f'<{bio_tag} : {word}> {txt}\n'
                    break
                else:
                    bio_tag = 'O'
            main_line = f'{i+1}\t{pos[i][0]}\t{pos[i][1]}\t{bio_tag}\n'
            # print(main_line)
            main_lines.append(main_line)

        for main_line in main_lines:
            if len(main_line.split('\t')[3]) > 2:
                word = main_line.split('\t')[1]
                tag = main_line.split('\t')[3].replace('\n', '')
                txt = txt.replace(word, f'<{word}:{tag}>')
                second_line = f'${txt}\n'
            else:
                second_line = f'${txt}\n'

        sentence_line = first_line + second_line
        # print(sentence_line)

        for main_line in main_lines:
            sentence_line += main_line

        sentence_line = sentence_line + '\n'
        sentence_lines.append(sentence_line)
```

```python
def process_txt(txt):
    pos = komoran.pos(txt)
    main_lines = []
    second_tags = []
    second_words = []

    for pos_word, pos_mor in pos:
        bio_tag = 'O'
        for ner_list in ner_lists:
            word, mor, tag = ner_list.split('\t')
            if pos_word == word and len(pos_word) == len(word) and pos_mor == mor:
                bio_tag = tag
                second_tags.append(tag)
                second_words.append(word)
                break

        main_line = f'{pos_word}\t{pos_mor}\t{bio_tag}\n'
        main_lines.append(main_line)

    first_line = f'; {txt}\n'
    for i in range(len(second_words)):
        txt = txt.replace(second_words[i], f'<{second_words[i]}:{second_tags[i]}>')
    second_line = f'${txt}\n'
    sentence_line = first_line + second_line + ''.join(main_lines) + '\n'
    return sentence_line

sentence_lines = []
for txt in tqdm(txts):
    sentence_line = process_txt(txt)
    sentence_lines.append(sentence_line)
```
