# 데이터 불러오기
### 영화명 / 등급 / 장르 / 국적 / 감독 / 배우

```python
df = pd.read_csv('c:/2nd_project/Data/movie_data/[KOBIS] 박스오피스_영화목록(2003.01~2023.07).csv')
df
```

```python
mov_lists = []   # 영화명 리스트
rat_lists = []   # 등급 리스트
gen_lists = []   # 장르 리스트
nat_lists = []   # 국적 리스트
dir_lists = []   # 감독 리스트
act_lists = []   # 배우 리스트
```

### 영화명 분류

```python
print('전체 영화 개수 : ', len(df))
print('중복 영화명 개수 : ',len(df[df['영화명'].duplicated()]))
```

```python
mov_lists = df['영화명'].drop_duplicates().dropna().tolist()
print(len(mov_lists))
mov_lists
```

### 등급 분류

```python
rat_lists = df['등급'].drop_duplicates().dropna().tolist()
print(len(rat_lists))
rat_lists
```

```python
rat_sep_lists = []

for rat_list in rat_lists:
    rats = rat_list.split(',')
    # print(dirs)
    # if len(dirs) > 1:
    #     print(dirs)
    #     print(len(dirs))
    #     print('='*20)
    for rat in rats:
        if rat == '기타' or rat == '미정':
            pass
        else:
            rat_sep_lists.append(rat)
    # print('*'*50)
rat_sep_lists = set(rat_sep_lists)
print(len(rat_sep_lists))
rat_sep_lists
```

### 장르 분류

```python
gen_lists = df['장르'].drop_duplicates().dropna().tolist()
print(len(gen_lists))
gen_lists
```

```python
gen_sep_lists = []

for gen_list in gen_lists:
    gens = gen_list.split(',')
    # print(gens)
    # print('='*20)
    for gen in gens:
        gn = gen.split('/')
        # print(gn)
        # print('='*20)
        for g in gn:
            # print(g)
            adds = g.split('(')
            for add in adds:
                if add != '기타':
                    add = re.sub(r'[^\w\s]', '', add)   # 특수기호 제거
                    gen_sep_lists.append(add)
    # print('*'*50)

gen_sep_lists = set(gen_sep_lists)
print(len(gen_sep_lists))
gen_sep_lists
```

### 국적 분류

```python
nat_lists = df['대표국적'].drop_duplicates().dropna().tolist()
print(len(nat_lists))
nat_lists
```

```python
# 국적 이상값 제거
nat_lists = [value for value in nat_lists if value != '총괄(연감)']
nat_lists = [value for value in nat_lists if value != '기타']
len(nat_lists)
```

```python
# 국적 추가
nat_lists.extend(['대한민국', '우리나라'])
len(nat_lists)
```

```python
nat_lists[:3]
```

### 감독 분류

```python
dir_lists = df['감독'].drop_duplicates().dropna().tolist()
print(len(dir_lists))
dir_lists
```

```python
dir_sep_lists = []

for dir_list in dir_lists:
    dirs = dir_list.split(',')
    # print(dirs)
    # if len(dirs) > 1:
    #     print(dirs)
    #     print(len(dirs))
    #     print('='*20)
    for dir in dirs:
        adds = dir.split('(')
        for add in adds:
            add = re.sub(r'[^\w\s]', '', add)   # 특수기호 제거
            dir_sep_lists.append(add)
    # print('*'*50)
dir_sep_lists = set(dir_sep_lists)
print(len(dir_sep_lists))
dir_sep_lists
```

### 배우 분류

```python
act_lists = df['배우'].drop_duplicates().dropna().tolist()
print(len(act_lists))
act_lists
```

```python
act_sep_lists = []

for act_list in act_lists:
    acts = act_list.split(',')
    # print(dirs)
    # if len(dirs) > 1:
    #     print(dirs)
    #     print(len(dirs))
    #     print('='*20)
    for act in acts:
        adds = act.split('(')
        for add in adds:
            add = re.sub(r'[^\w\s]', '', add)   # 특수기호 제거
            act_sep_lists.append(add)
    # print('*'*50)
act_sep_lists = set(act_sep_lists)
print(len(act_sep_lists))
act_sep_lists
```

```python
# 배우 이상값 제거
act_sep_lists = [value for value in act_sep_lists if value != '추천']
act_sep_lists = [value for value in act_sep_lists if value != 'SS ']
len(act_sep_lists)
```
