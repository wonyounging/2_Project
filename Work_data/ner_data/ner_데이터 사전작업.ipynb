# 데이터 불러오기
### 영화명, 날짜/시간 데이터

```python
# 파일 경로
file_path = 'd:/2nd_project/Data/user_dic_add.txt'

# 파일 열기
with open(file_path, 'r', encoding='utf-8') as file:
    lines = file.readlines()

txts = []
# 파일 내용 줄 단위로 출력
for line in lines:
    txts.append(line.strip())

txts
```

### B_DT(날짜, 시간) / B_MOVIE(영화명) tag 추가

```python
for txt in txts:
    word = txt.split('\t')[0]
    pos = txt.split('\t')[1]
    # print(word)
    # print(pos)

    if (
        re.match(r'\b([0-9]|1[0-9]|2[0-4])시\b', word) or        # 시에 tag : B_DT 추가
        re.match(r'\b([0-9]|[0-5][0-9]|60)분\b', word) or               # 분에 tag : B_DT 추가
        re.match(r'\b([1-9]|1[0-2])월\b', word) or                      # 월에 tag : B_DT 추가
        re.match(r'\b([1-9]|1[0-9]|2[0-9]|3[0-1])일\b', word) or        # 일에 tag : B_DT 추가
        re.match(r'\b(월|화|수|목|금|토|일)요일\b', word)               # 요일에 tag : B_DT 추가
        ) and len(word) <= 3:
        # print(word)
        info = word+'\t'+pos+"\tB_DT"
        ner_tag.append(info)
    else:
        info = word+'\t'+pos+"\tB_MOVIE"        # default B_MOVIE
        ner_tag.append(info)
```

# 데이터 불러오기
### 등급 / 장르 / 국적 / 감독 / 배우

```python
df = pd.read_csv('d:/2nd_project/Data/movie_data/2023_01~07_movie.csv')
df
```

```python
rat_lists = []   # 등급 리스트
gen_lists = []   # 장르 리스트
nat_lists = []   # 국적 리스트
dir_lists = []   # 감독 리스트
act_lists = []   # 배우 리스트
```

### 등급 분류

```python
rat_lists = df['등급'].drop_duplicates().dropna().tolist()
print(len(rat_lists))
rat_lists
```

### 장르 분류

```python
gen_lists = df['장르'].drop_duplicates().dropna().tolist()
print(len(gen_lists))
gen_lists
```

```python
gen_sep_lists = []

for gen_list in gen_lists:
    gens = gen_list.split(',')
    # print(gens)
    # print('='*20)
    for gen in gens:
        gn = gen.split('/')
        # print(gn)
        # print('='*20)
        for g in gn:
            # print(g)
            gen_sep_lists.append(g)
    # print('*'*50)

gen_sep_lists = set(gen_sep_lists)
print(len(gen_sep_lists))
gen_sep_lists
```

### 국적 분류

```python
nat_lists = df['대표국적'].drop_duplicates().dropna().tolist()
print(len(nat_lists))
nat_lists
```

```python
dir_lists = df['감독'].drop_duplicates().dropna().tolist()
print(len(dir_lists))
dir_lists
```

```python
dir_sep_lists = []

for dir_list in dir_lists:
    dirs = dir_list.split(',')
    # print(dirs)
    # if len(dirs) > 1:
    #     print(dirs)
    #     print(len(dirs))
    #     print('='*20)
    for dir in dirs:
        dir_sep_lists.append(dir)
    # print('*'*50)
dir_sep_lists = set(dir_sep_lists)
print(len(dir_sep_lists))
dir_sep_lists
```

### 배우 분류

```python
act_lists = df['배우'].drop_duplicates().dropna().tolist()
print(len(act_lists))
act_lists
```

```python
act_sep_lists = []

for act_list in act_lists:
    acts = act_list.split(',')
    # print(dirs)
    # if len(dirs) > 1:
    #     print(dirs)
    #     print(len(dirs))
    #     print('='*20)
    for act in acts:
        act_sep_lists.append(act)
    # print('*'*50)
act_sep_lists = set(act_sep_lists)
print(len(act_sep_lists))
act_sep_lists
```

### B_RAT(등급) tag 추가

```python
for rat_list in rat_lists:
    print(rat_list)
    info = rat_list+'\tNNG\tB_RAT'
    ner_tag.append(info)
```

### B_NAT(국적) tag 추가

```python
for nat_list in nat_lists:
    print(nat_list)
    info = nat_list+'\tNNG\tB_NAT'
    ner_tag.append(info)
```

### B_GEN(장르) tag 추가

```python
gen_sep_lists
for gen_sep_list in gen_sep_lists:
    print(gen_sep_list)
    info = gen_sep_list+'\tNNG\tB_GEN'
    ner_tag.append(info)
```

### B_DIR(감독) tag 추가

```python
dir_sep_lists
for dir_sep_list in dir_sep_lists:
    print(dir_sep_list)
    info = dir_sep_list+'\tNNG\tB_DIR'
    ner_tag.append(info)
```

### B_ACT(배우) tag 추가

```python
act_sep_lists
for act_sep_list in act_sep_lists:
    print(act_sep_list)
    info = act_sep_list+'\tNNG\tB_ACT'
    ner_tag.append(info)
```
