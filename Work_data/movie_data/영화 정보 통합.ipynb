# 영화 데이터 하나로 만들기
[KOBIS] 박스오피스_영화목록(2003.01~2022.12) + [KOBIS] 박스오피스_영화목록(2023.01~2023.07)

```python
import pandas as pd
```

[KOBIS] 박스오피스_영화목록(2003.01~2022.12)

```python
df1 = pd.read_csv('c:/2nd_project/Data/movie_data/[KOBIS] 박스오피스_영화목록(2003.01~2022.12).csv')
df1.rename(columns={'배우 ': '배우'}, inplace=True)
df1
```

[KOBIS] 박스오피스_영화목록(2023.01~2023.07)

```python
df2 = pd.read_csv('c:/2nd_project/Data/movie_data/[KOBIS] 박스오피스_영화목록(2023.01~2023.07).csv')
df2
```

```python
# 중복된 행 확인
duplicates = df1[df1.duplicated() | df2.duplicated()]
print("중복된 행:")
print(duplicates)
```

```python
print(df1.columns)
print(df2.columns)
```

```python
df = pd.concat([df1, df2], ignore_index=True)
df
```

결측값 확인

```python
df['영화명'].isnull().sum()
```

```python
df[df['영화명'].duplicated(keep=False)].sort_values(by='영화명')
```

```python
df.drop_duplicates(subset=['영화명', '개봉일'], inplace=True)
df
```

```python
same_name_list = []
same_name_list = df[df['영화명'].duplicated(keep=False)].sort_values(by='영화명')
same_name_list
```

# 데이터 저장

```python
same_name_list.to_csv('c:/2nd_project/Data/movie_data/동일명 영화.csv', index=None)
```

```python
df.to_csv('c:/2nd_project/Data/movie_data/[KOBIS] 박스오피스_영화목록(2003.01~2023.07).csv', index=None)
```
