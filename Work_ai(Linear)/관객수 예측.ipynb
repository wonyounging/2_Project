```python
import pandas as pd
```

```python

df = pd.read_csv('c:/2nd_project/Data/movie_data/2003~2022 일별.csv')
# df = pd.read_csv('c:/2nd_project/Data/movie_data/2023년 1~7월/2023(01~07)_merge_mid.csv')
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.groupby('영화명').apply(lambda x: x.sort_values('기준일')).reset_index(drop=True)
df
```

```python
df_test = df[df['영화명'] == '극한직업']
df_test
```

```python
df_test_watch = df_test['관객수']
df_test_watch
```

```python
df_test_watch.plot()
```

```python
import matplotlib.pyplot as plt
#1차 차분을 통해 데이터를 정상화하는 과정

diff1 = df_test_watch.diff(periods=1)
diff1.plot()

plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller
import numpy as np

#로그 후 차분한 자료를 adfuller 함수로 안정적인 시계열인지 확인
diff1 = df_test_watch.diff(periods=1).iloc[1:]
#        로그                  1차차분

result = adfuller(diff1)
#         안정적인 시계열 데이터 여부

print(result)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
```

```python
from pmdarima.arima import auto_arima
model = auto_arima(df_test_watch)
model
```

```python
#학습용,검증용 구분
# train = df_test_watch[7:] # 학습용
# test = df_test_watch[:-10]  # 검증용

train = df_test_watch[:20] # 학습용
test = df_test_watch[20:]  # 검증용

print(len(df_test_watch))
print(df_test_watch.values)
```

```python
#arima(1,2,1)을 수행할 경우  order=c(1,2,1)로 모형을 생성함
from statsmodels.tsa.arima.model import ARIMA
model = ARIMA(train, order=(5,1,1)).fit()
#       시계열 모형
model.plot_diagnostics(figsize=(16, 8))

plt.show()
```

```python
predictions = model.predict(start=1, end=len(test), dynamic=False)
predictions.index=len(train)+(predictions.index-1)
predictions
```

```python
df_test_watch.iloc[30]
```

```python
df_test_watch.reset_index(drop=True, inplace=True)
df_test_watch
```

```python
plt.plot(df_test_watch,label='Real')
plt.plot(predictions,label='Predict')
plt.legend(loc='upper left',fontsize=8)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error  

mse = mean_squared_error(test, predictions)  
#                       실제    모형출력

rmse = np.sqrt(mse)  # 평균제곱근오차
rmse
```

```python
# 주어진 시계열 데이터
data = df_test_watch[:70]  # 데이터를 직접 입력해주세요

# ARIMA 모델 훈련
order = (5, 1, 1)  # ARIMA 모델의 차수 설정
model = ARIMA(data, order=order)
model_fit = model.fit()

# 예측 시작
forecast_steps = 50  # 15일 이후를 예측
forecast = model_fit.forecast(steps=forecast_steps)

# 시계열 데이터와 예측 결과 그래프로 표현
plt.figure(figsize=(10, 6))
plt.plot(data, label='Original Data')
plt.plot(np.arange(len(data), len(data) + forecast_steps), forecast, label='Forecast')
plt.legend()
plt.title('ARIMA Forecast')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
