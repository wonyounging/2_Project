```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from tqdm import tqdm
import pandas as pd

# NaN 값 처리를 위한 함수 정의
def preprocess_text(text):
    return str(text) if not pd.isnull(text) else ""

# TfidfVectorizer로 문장을 벡터화
vectorizer = TfidfVectorizer()

# 각 행에 대해 코사인 유사도 계산 및 업데이트
def calculate_cosine_similarity(row):
    try:
        tfidf_matrix = vectorizer.fit_transform([all_story, row['줄거리']])  # TfidfVectorizer로 문장을 벡터화
        cosine_sim = cosine_similarity(tfidf_matrix[0], tfidf_matrix[1])    # 코사인 유사도 계산
        return cosine_sim[0][0]
    except:
        return 0.0
```

```python
# 코사인 유사도 계산 및 업데이트
df['줄거리'] = tqdm(df.apply(calculate_cosine_similarity, axis=1))
```

```python
df.columns = ['Rating', 'Genre', 'Nation', 'Director', 'Actor', 'Story', 'Score']
df
```

```python
df = df[['Rating', 'Genre', 'Nation', 'Story', 'Score']]
df
```

```python
data = pd.get_dummies(df, columns=['Rating', 'Genre', 'Nation'])
data.head(3)
```

```python
X = data.drop(columns=['Score'])
y = data['Score']
print(X.shape)
print(y)
```

```python
from sklearn.model_selection import train_test_split
# 학습 데이터와 테스트 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
# 선형 회귀 모델 생성 및 학습
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error
# 모델 평가
y_pred = model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = mse ** 0.5
print(f'평균 제곱 오차 (MSE): {mse}')
print(f'평균 제곱근 오차 (RMSE): {rmse}')
```

```python
# 모델을 사용하여 새로운 데이터에 대한 예측 수행
new_data = pd.DataFrame({'rating_0': [1], 'rating_11': [0], 'genre_1': [0], 'genre_2': [1], 'nation_1': [0], 'nation_2': [1], 'nation_3': [0], 'story': [4]})
predicted_score = model.predict(new_data)
print(f'예측된 평점: {predicted_score[0]}')
```

```python
import tomotopy as tp
model = tp.LDAModel(k=5, min_cf=2)
```

```python
from konlpy.tag import Komoran
komoran = Komoran(userdic='c:/2nd_project/Data/[최종] user_dic.txt')
n = komoran.nouns(text)
n2 = [x for x in n if len(x)>1]
if len(n2)>0:
    model.add_doc(n2)
```

```python
model.train(0)

print('Total docs:', len(model.docs))
print('Total words:', model.num_words)
```

```python
#200회 학습
model.train(200)

test_list = []
# 학습된 토픽들을 출력
for i in range(model.k):
    # 토픽별 상위 단어 10개 추출
    res = model.get_topic_words(i, top_n=10)

    print(f'Topic #{i}', end='\t')
    print(', '.join(w for w, p in res))
    test_list.append(', '.join(w for w, p in res))
```

```python
t = []
for i in test_list:
    for j in i.split(','):
        t.append(j.strip())
t

from collections import Counter
# 단어 빈도 계산
word_count = Counter(t)

# 가장 많이 나온 단어 3개 추출
most_common_words = word_count.most_common(3)

# 결과 출력
for word, count in most_common_words:
    print(f'{word}: {count}번')
```
