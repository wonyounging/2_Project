### 불필요 컬럼 제거

```python
# ID	ID  (의미 X)
# LCLAS_NM	대분류명    (장소 단일값 제외)
# MLSFC_NM	중분류명    (문화시설 단일값 제외)
# POI_ID	POI_ID  (의미 X)
# POI_NM	POI명
# BHF_NM	지점명
# ASSTN_NM	보조명  (결측값)
# CL_CD	분류코드    (분류명과 같은 의미값)
# CL_NM	분류명  (분류명과 같은 의미값)
# PNU	지번고유코드    (의미 X)
# CTPRVN_NM	시도명
# SIGNGU_NM	시군구명
# LEGALDONG_NM	법정동명
# LI_NM	리명
# LNBR_NO	번지번호
# LEGALDONG_CD	법정동코드  (의미 X)
# ADSTRD_CD	행정동코드  (의미 X)
# RDNMADR_CD	도로명주소코드  (의미 X)
# RDNMADR_NM	도로명주소명
# BULD_NO	건물번호
# LC_LO	위치경도    (보류)
# LC_LA	위치위도    (보류)
# GID_CD	격자코드    (보류)
# LAST_CHG_DE	최종변경일자    (단일값)
# ORIGIN_NM	출처명  (단일값)
# FILE_NM	파일명  (단일값)
# BASE_DE	기준일자    (단일값)


# POI_NM	POI명
# BHF_NM	지점명
# CTPRVN_NM	시도명
# SIGNGU_NM	시군구명
# LEGALDONG_NM	법정동명
# LI_NM	리명
# LNBR_NO	번지번호
# RDNMADR_NM	도로명주소명
# BULD_NO	건물번호
# LC_LO	위치경도    (보류)
# LC_LA	위치위도    (보류)
# GID_CD	격자코드    (보류)
```

```python
df = df[['POI_NM', 'BHF_NM', 'CTPRVN_NM', 'SIGNGU_NM', 'LEGALDONG_NM', 'LI_NM', 'LNBR_NO', 'RDNMADR_NM', 'BULD_NO']]
df
```

### 컬럼명 변경

```python
df.columns = ['영화관명', '지점명', '시도명', '시군구명', '법정동명', '리명','번지번호', '도로명주소명', '건물번호']
df.reset_index(drop=True, inplace=True)
df
```

### 이상 데이터 정리

```python
df.loc[(df['영화관명'] == 'CGV') & (df['지점명'] == '경산점'), '건물번호'] = '6-5'
df.loc[(df['영화관명'] == 'CGV') & (df['지점명'] == '울산진장점'), '건물번호'] = '51-10'
df.loc[(df['영화관명'] == 'CGV') & (df['지점명'] == '광주금남로'), '건물번호'] = '16-7'
df.loc[(df['영화관명'] == 'CGV') & (df['지점명'] == '청주지웰시티점'), '건물번호'] = '47-1'
df.loc[(df['영화관명'] == '롯데시네마') & (df['지점명'] == '양주고읍점'), '건물번호'] = '6-16'
df.loc[(df['영화관명'] == '롯데시네마') & (df['지점명'] == '향남점'), '건물번호'] = '7-21'
df.loc[(df['영화관명'] == '롯데시네마') & (df['지점명'] == '천안점'), '건물번호'] = '67-8'
df.loc[(df['영화관명'] == '메가박스') & (df['지점명'] == '송도점'), '건물번호'] = '33-4'
df.loc[(df['영화관명'] == '메가박스') & (df['지점명'] == '광주상무점'), '건물번호'] = '21-6'
df.loc[(df['영화관명'] == '메가박스') & (df['지점명'] == '양주점'), '건물번호'] = '6-6'
df.loc[(df['영화관명'] == '메가박스') & (df['지점명'] == '오창점'), '건물번호'] = '8-9'
```
